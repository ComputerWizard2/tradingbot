# æ¸…æ´—æ•°æ®

```python
import pandas as pd

# ========= 1ï¸âƒ£ è¯»å–åŸå§‹æ•°æ® =========
raw_path = "data/XAU_1d_data.csv"   # æ”¹æˆä½ çš„è·¯å¾„
clean_path = "data/xauusd_d1_from_m1.csv"

print("ğŸ“¥ è¯»å–åŸå§‹æ•°æ®...")
df = pd.read_csv(raw_path, sep=';')

print("åŸå§‹åˆ—åï¼š", df.columns.tolist())
print("åŸå§‹æ•°æ®é‡ï¼š", len(df))

# ========= 2ï¸âƒ£ æ ‡å‡†åŒ–åˆ—å =========
df.columns = df.columns.str.strip()  # å»é™¤ç©ºæ ¼

# ç»Ÿä¸€é‡å‘½å
df = df.rename(columns={
    'Date': 'time',
    'Open': 'open',
    'High': 'high',
    'Low': 'low',
    'Close': 'close',
    'Volume': 'volume'
})

# ========= 3ï¸âƒ£ æ—¶é—´æ ¼å¼è½¬æ¢ =========
df['time'] = pd.to_datetime(df['time'])

# ========= 4ï¸âƒ£ æ’åº =========
df = df.sort_values('time')

# ========= 5ï¸âƒ£ åˆ é™¤é‡å¤ =========
before = len(df)
df = df.drop_duplicates(subset='time')
after = len(df)

print(f"ğŸ§¹ åˆ é™¤é‡å¤: {before - after}")

# ========= 6ï¸âƒ£ æ£€æŸ¥ç¼ºå¤± =========
print("ğŸ” ç¼ºå¤±å€¼ç»Ÿè®¡ï¼š")
print(df.isna().sum())

# ========= 7ï¸âƒ£ æ£€æŸ¥æ•°æ®ç±»å‹ =========
print("\næ•°æ®ç±»å‹ï¼š")
print(df.dtypes)

# ========= 8ï¸âƒ£ ä¿å­˜ =========
df = df[['time','open','high','low','close','volume']]
df.to_csv(clean_path, index=False)

print("\nâœ… æ¸…æ´—å®Œæˆï¼")
print("ä¿å­˜è·¯å¾„ï¼š", clean_path)
print("æœ€ç»ˆæ•°æ®é‡ï¼š", len(df))
print(df.head())
```

## 6ï¸âƒ£ Start Training!

### Training Configurations:

**Quick Test (10 minutes):**
- Steps: 10,000
- Good for testing the pipeline

**Short Training (1-2 hours):**
- Steps: 100,000
- Will show initial learning

**Full Training (5-7 hours on A100, 12-15 hours on T4):**
- Steps: 1,000,000
- Production-ready model

```python
# Choose your training configuration
TRAINING_STEPS = 1_000_000  # Change to 10_000 for quick test, 100_000 for short training
BATCH_SIZE = 128 if device == 'cuda' else 16  # Larger batch on GPU

print(f"ğŸ‹ï¸  Training Configuration:")
print(f"   â€¢ Steps: {TRAINING_STEPS:,}")
print(f"   â€¢ Batch size: {BATCH_SIZE}")
print(f"   â€¢ Device: {device}")
print(f"\nâ±ï¸  Estimated time:")

if device == 'cuda':
    gpu_name = torch.cuda.get_device_name(0)
    if 'A100' in gpu_name:
        time_estimate = TRAINING_STEPS / 1_000_000 * 5  # ~5 hours for 1M steps on A100
    elif 'V100' in gpu_name:
        time_estimate = TRAINING_STEPS / 1_000_000 * 8  # ~8 hours for 1M steps on V100
    else:  # T4
        time_estimate = TRAINING_STEPS / 1_000_000 * 12  # ~12 hours for 1M steps on T4
    print(f"   â€¢ ~{time_estimate:.1f} hours on {gpu_name}")
else:
    print(f"   â€¢ CPU training will be VERY slow (not recommended)")

print(f"\nâš ï¸  Make sure Colab stays open (or use Colab Pro+ for background execution)")
```
